```python
import matplotlib.pyplot as plt


textract_values = [
    94, 57, 32, 54, 57, 73, 67, 57, 74, 63,
    68, 57, 72, 78, 75, 48, 54, 69
]

tesseract_values = [
    84, 54, 32, 61, 54, 47, 85, 49, 60, 76,
    76, 48, 87, 32, 66, 50, 71, 60
]


documents = list(range(1, len(textract_values) + 1))


plt.figure(figsize=(12, 6))
plt.plot(documents, textract_values, label='Textract', color='blue', marker='o')
plt.plot(documents, tesseract_values, label='Tesseract', color='orange', marker='o')


plt.title('Comparação entre Textract e Tesseract')
plt.xlabel('Número de Documentos')
plt.ylabel('Número de Palavras Válidas')
plt.xticks(documents)
plt.legend()
plt.grid()

plt.show()
```
